# Librerie

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DatasetCyclists = pd.read_csv("dataset/cyclists.csv")
DatasetRace = pd.read_csv("dataset/races.csv")
```

# **Dataset cyclists distribution analysis**

```python
PlotDatasetCyclists = pd.DataFrame({
    "Elements": list(DatasetCyclists.keys()),
    "Total values": [DatasetCyclists.shape[0]]*len(list(DatasetCyclists.keys())),
    "Missing values": list(DatasetCyclists.isnull().sum()),
    "Usable values": list(DatasetCyclists.shape[0] - DatasetCyclists.isnull().sum())
})

PlotDatasetCyclists.set_index('Elements', inplace=True)
PlotDatasetCyclists.plot(kind='bar', figsize=(10, 6))

plt.title('Dataset cyclists distrbution')
plt.xlabel('Features')
plt.ylabel('Valori')
plt.show()
```

```python
InterestingFeatures = ["birth_year", "weight", "height"]
PlotsLambda = [lambda f, ax, i: ax[i].hist(DatasetCyclists[f], bins = 10, alpha = 0.7, color = "blue")]

i = 0

fig, ax = plt.subplots(len(InterestingFeatures), len(PlotsLambda))

for feature in InterestingFeatures:  
    ax[i].set_title(feature) 
    for plot in PlotsLambda:    
        plot(feature, ax, i)
        i += 1

#plt.title("Distribution cyclists dataset")
plt.tight_layout()  # Per evitare sovrapposizioni
plt.show()
```
