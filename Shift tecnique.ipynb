```python
import random
import matplotlib.pyplot as plt
import numpy as np

Points = np.array([[random.random(), random.random()] for i in range(100)])

def Plot(Points):
    
    points = list(Points)

    X = [i[0] for i in points]
    Y = [i[1] for i in points]
    
    plt.scatter(X, Y, c = "red")
    plt.show()
    
def DistanceMeasure(x, y):
    return np.sum((x - y)**(0.5))

def ShiftMethod(points, k, Aggregation):
    return np.array([Aggregation(sorted(points.copy(), key = lambda x: DistanceMeasure(p, x))[:k]) for p in points])

def MOD(points, k):
    return ShiftMethod(points, k, lambda neighbors: np.mean(neighbors, axis = 0))

def DOD(points, k):
    return ShiftMethod(points, k, lambda neighbors: min(neighbors, key = lambda x: sum(DistanceMeasure(x, y) for y in neighbors)))
```

## MOD

```python
PointsMOD = Points.copy()

Plot(PointsMOD)

for i in range(10):
    
    PointsMOD = MOD(PointsMOD, 3)
    Plot(PointsMOD)
```

## DOD

```python
PointsDOD = Points.copy()

Plot(PointsDOD)

for i in range(10):
    
    PointsDOD = DOD(PointsDOD, 3)
    Plot(PointsDOD)
```
